### Analysis: Price = f(S,K,T,sigma,r)

```python
S, K, T, sigma, r = 100, 110, 2, 0.2, 0.03
simple_pricer_function = bs_tf.Pricer.plain_vanilla_closed_formula_with_2nd_greeks(enable_greeks=False)
```

```python
# Fixed:
S, K, T, sigma = 100, 110, 2, 0.2

# Variable: r
rs = np.linspace(-0.5,2,30)

# Generate
ss = [S for _ in rs]
ks = [K for _ in rs]
ts = [T for _ in rs]
sigmas = [sigma for _ in rs]

# Create dataframe and title for plot
df = pd.DataFrame(data={'S':ss, 'K':ks, 'T':ts, 'sigma':sigmas, 'r':rs, 'Price': 0})
title = "Variable: r, Fixed: " + ", ".join(list(map(lambda ls: ls[0]+ls[1], zip(['S:', 'K:', 'T:', 'sigma:'], list(map(str, [S, K, T, sigma]))))))

# Apply B&S Formula
def func(row):
    return simple_pricer_function(*row)[0]
df.Price = df[['S', 'K', 'T', 'sigma', 'r']].apply(func, 1)

# Plot
fig, ax = plt.subplots(figsize=(10, 5), ncols=1, nrows=1)
sns.lineplot(x='r', y='Price', data=df, ax=ax)
ax.set_title(title)
```
