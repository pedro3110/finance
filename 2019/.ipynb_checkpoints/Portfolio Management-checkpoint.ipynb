```python
def generate_dataframe(ticker, init_date, end_date):
    df = pd.DataFrame(columns=['date'])
    yahoo_financials = YahooFinancials(ticker)
    history = yahoo_financials.get_historical_price_data(init_date, end_date, 'daily')
    
    df['date'] = [x['formatted_date'] for x in history[ticker]['prices']]
    df['date'] = pd.to_datetime(df['date'])
    adjclose = [x['adjclose'] for x in history[ticker]['prices']]
    df[ticker] = adjclose
    return df
```

```python
fixed_income = [
    {'name': 'Treasury Yield 10 Years', 'ticker': '^TNX'},
    {'name': 'Treasury Yield 30 Years', 'ticker': '^TYX'},
#     {'name': 'iShares iBoxx $ High Yield Corporate Bond ETF', 'ticker': 'HYG'}
    {'name': 'iShares iBoxx $ Investment Grade Corporate Bond ETF', 'ticker': 'LQD'}
]
equity = [
    {'name': 'S&P', 'ticker': '^GSPC'},
    {'name': 'Vanguard FTSE Emerging Markets', 'ticker': 'VWO'}
]
```

```python
init_date = '2016-10-17'
end_date = '2019-10-19'

# Evolucion
for i, f in enumerate(fixed_income):
    ticker = f['ticker']
    name = f['name']
    if i == 0:
        df = generate_dataframe(ticker, init_date, end_date)
    else:
        new_df = generate_dataframe(ticker, init_date, end_date); new_df = new_df.drop('date', 1)
        df = pd.concat([df, new_df], axis=1)

# Returns
df_ret = df.copy()
for col in df_ret.columns.tolist():
    if col != 'date':
        df_ret[col] = df_ret[col].diff() / df_ret[col]
# Returns Acumulado
df_ret_acum = df_ret.tail(n=df_ret.shape[0]-1).copy()
for col in df_ret.columns.tolist():
    if col != 'date':
        df_ret_acum[col] = df_ret_acum[col] + 1
        df_ret_acum[col] = df_ret_acum[col].cumprod()
```

### Maximizacion (encontrar weights para maximizar retorno esperado; usando retornos esperados de cada asset class)

```python
df_ret.head()
```

```python
# Retornos esperados
# ASSET_CLASSES = ['AAPL', 'MSFT', 'INTC']
# μ = df_ret[['AAPL', 'MSFT', 'INTC']].describe().T[['min', 'mean', 'std', 'max']]['mean'].tolist()

ASSET_CLASSES = ['^TNX', '^TYX', 'LQD']
μ = df_ret[ASSET_CLASSES].describe().T[['min', 'mean', 'std', 'max']]['mean'].tolist()


Σ = df_ret[ASSET_CLASSES].corr().values

w = np.ones(Σ.shape[0])
σ = lambda w: w.dot(Σ.dot(w.T))

# Constraints
w_constraint = ({'type': 'eq', 'fun': lambda w: sum(w) - 1.})
r_constraint = ({'type': 'eq', 'fun': lambda w: np.dot(μ, w) - r_expected})
w_bound      = []
w_bound      = [(0, 1) for _ in μ]

# Optimization
r_expected = 0.01
w_opt = optimize.minimize(σ, w, method = 'SLSQP', 
                                constraints = [w_constraint, r_constraint], 
                                bounds = w_bound)
print('Sigma: ', σ(w_opt.x))
print('Return: ', w_opt.x * μ)
print('Weights: ' , [np.round(x, 5) for x in w_opt.x], '=>', sum(w_opt.x))
```
