```python
# Optimizer para hacer CONSTANT REBALANCING
def get_f(cost):
    def f(w):
        tot = np.array([1 for _ in s])
        for i in z_diff.index.values:
            alloc = z_diff.iloc[i].values
            ret = w * alloc
            # new weight
            w_1 = np.array([x/sum(ret) for x in ret])
            # rebalancing of weight
            d = w - w_1
            # cost of rebalancing
            c = cost * d
            # total value after rebalancing
            tot = tot + ret - c
        return - sum(tot)
    return f
```

```python
# Params
min_date = df_ret_acum.date.min()
max_date = df_ret_acum.date.max()
dates = ['2017-05-01', '2018-05-07', max_date]

# optimize
z = df_ret_acum[df_ret_acum.date.isin(dates)].reset_index().drop('index', 1)
z_diff = z.diff()
z_diff = z_diff.tail(n=z_diff.shape[0]-1).reset_index().drop('index', 1)
```

```python
w_constraint = ({'type': 'eq', 'fun': lambda w: sum(w) - 1.})
w_bound      = [(0, 1) for _ in s]

# methods = ['Nelder-Mead', 'Powell', 'CG', 'BFGS', 'Newton-CG', 'L-BFGS-B', 'TNC', 'COBYLA',
#     'SLSPQ', 'trust-constr', 'dogleg', 'trust-ncg', 'trust-krylov', 'trust-exact']
method = 'SLSQP'

w_0 = [1./len(s) for _ in s]
cost = np.array([1,2,1,1,1]) / 100
w_opt = optimize.minimize(get_f(cost), w_0, method = method, 
                          constraints = [w_constraint], 
                          bounds = w_bound)
w_opt
```
